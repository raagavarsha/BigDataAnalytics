```python
import matplotlib.pyplot as plt
```

```python
df[['magnitude', 'depth', 'cdi', 'mmi', 'sig']].hist(bins=20, figsize=(12,8))
plt.suptitle("Distributions of Earthquake Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df[['magnitude', 'depth', 'cdi', 'mmi', 'sig']].plot(kind='box')
plt.title("Boxplots of Numerical Features")
plt.show()
```

```python
df['alert'].value_counts().plot(kind='bar', color='orange')
plt.title("Number of Earthquakes by Alert Level")
plt.xlabel("Alert Level")
plt.ylabel("Count")
plt.show()
```

```python
correlation = df[['magnitude', 'depth', 'cdi', 'mmi', 'sig']].corr()
plt.figure(figsize=(6,5))
plt.matshow(correlation, fignum=1)
plt.colorbar()
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.title("Correlation Matrix", pad=20)
plt.show()
```

```python
import itertools
```

```python
cols = ['magnitude', 'depth', 'cdi', 'mmi', 'sig']
for x, y in itertools.combinations(cols, 2):
    plt.figure(figsize=(5,4))
    plt.scatter(df[x], df[y], alpha=0.5)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'{x} vs {y}')
    plt.show()
```
